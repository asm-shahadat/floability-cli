# Floability Makeflow Backpack

This notebook gives an example of how to use Floability to package up a workflow written using the [Makeflow](http://ccl.cse.nd.edu/software/makeflow) workflow system.  This backpack directory contains all the elements needed to run the workflow:
- [notebook.ipynb](notebook.ipynb) is this file, the notebook interface to the workflow.
- [workflow.jx](workflow.jx) describes the workflow itself in the JX format.
- [environment.yml](environment.yml) describes the software needed, in this case the imagemagick software
- [data.yml](data.yml) describes the data file that the workflow is expecting as input.

# Running the Workflow Locally

Once deployed using the `floability` command, select the next cell and execute with shift-return.  It will invoke Makeflow to run the workflow directly as a child process:


```python
!makeflow -T local --jx workflow.jx
```

# Cleaning the Workflow

After running the workflow, the output files will be left in the working directory.  Use the `--clean` option of `makeflow` to remove all the files that the workflow generated:

```python
!makeflow --jx workflow.jx --clean
```

# Running the Workflow on the Cluster

Now run the workflow again, this time sending tasks to TaskVine workers running on the cluster.(The `floability` command should have set up those workers for you.)

```python
!makeflow -T local -N $VINE_MANAGER_NAME --jx workflow.jx
```
